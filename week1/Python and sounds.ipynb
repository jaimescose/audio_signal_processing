```python
from scipy.io import wavfile
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Reading a wav file
fs, x = wavfile.read('Melody1 150bpm.wav')
print(f'{fs} Hz, duration {len(x)/fs:.2f} seconds')

# this numpy arry has a shape of (N, 2) because it's a stereo file
print(x.shape)

# Plotting a wav file
plt.plot(x)
plt.show()
```

```python
# Ploting the wav file using time axis in seconds
t = np.arange(len(x))/fs

plt.plot(t,x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Plotting the wav file using time axis in seconds')
plt.show()
```

```python
# writing wav files
tail = x[600000:]
wavfile.write('tail.wav', fs, tail)

plt.plot(tail)
plt.show()
```
