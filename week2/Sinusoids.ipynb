```python
from matplotlib import pyplot as plt
import numpy as np
```

## Real sinewave

```python
# Define the sinusoidal wave
A = .8 # Amplitude
f0 = 1000 # Frequency
phi = np.pi/2 # Phase
fs = 44100 # Sampling frequency
t = np.arange(-.002, .002, 1/fs) 

# Generate the wave
x = A *np.cos(2* np.pi * f0 * t + phi)
```

```python
# Plotting a sinusoidal signal
plt.plot(t, x)
plt.axis([-.002, .002, -.8, .8])
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.show()
```

## Complex sinewave

```python
N = 500 # number of samples
k = 3 # frequency
n = np.arange(-N/2, N/2) # time axis

s = np.exp(1j * 2 * np.pi * k * n / N) # complex exponential
```

### Plotting the real part of the copmplex sinusoidal

```python
plt.plot(n, s.real)
plt.axis([-N/2, N/2, -1, 1])
plt.xlabel('n')
plt.ylabel('amplitud')

# we will observe 3 periods of the signal (k=3)
plt.show()
```

### Plotting the imaginary part of the copmplex sinusoidal

```python
plt.plot(n, s.imag)
plt.axis([-N/2, N/2, -1, 1])
plt.xlabel('n')
plt.ylabel('amplitud')

# we will observe 3 periods of the signal (k=3)
plt.show()
```
