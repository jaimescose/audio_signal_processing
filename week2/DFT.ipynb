```python
import matplotlib.pyplot as plt
import numpy as np
```

## DFT Calculation

```python
N = 64 # number of samples
k0 = 7 # frequency
n_array = np.arange(N)
x = np.exp(1j * 2 * np.pi * k0 * n_array / N) # complex exponential to calculate the samples of the signal
```

```python
X = np.array([])

for k in range(N):
    # we compute the complex exponential at every frequency (k)
    # the complex exponential is not expressed as negative, so we use conjugate it later on
    s = np.exp(1j * 2 * np.pi * k * n_array / N)

    # we compute the output spectrum
    # so, we append the inner product of the signal and the complex exponential
    X = np.append(X, sum(x * np.conjugate(s)))
```

```python
# plotting the spectrtrum of the signal
plt.plot(n_array, abs(X))
plt.axis([0, N - 1, 0, N])
plt.xlabel('n')
plt.ylabel('amplitud')

plt.show()
```

### Real signal

```python
N = 64 # number of samples
k0 = 7.5 # frequency

x = np.cos(2 * np.pi * k0 * n_array / N)

# compute it again using a real signal (cosine)
X = np.array([])
nv = np.arange(-N/2, N/2)
kv = np.arange(-N/2, N/2)

for k in kv:
    # we compute the complex exponential at every frequency (k)
    # the complex exponential is not expressed as negative, so we use conjugate it later on
    s = np.exp(1j * 2 * np.pi * k * nv / N)

    # we compute the output spectrum
    # so, we append the inner product of the signal and the complex exponential
    X = np.append(X, sum(x * np.conjugate(s)))
```

```python
# plotting the spectrtrum of the signal (real signal is the sum of two complex exponentials)
plt.plot(kv, abs(X))
plt.axis([-N/2, N/2 - 1, 0, N])
plt.xlabel('n')
plt.ylabel('amplitud')

# output should be around 7 and -7
plt.show()
```

## Inverse DFT calculation

```python
y = np.array([])

for n in nv:
    s = np.exp(1j * 2 * np.pi * kv * n / N)
    y = np.append(y, sum(X * s))

y = y / N

plt.plot(nv, y)
plt.axis([-N/2, N/2, -1, 1])

plt.show()
```
